```python
import pandas as pd

vents = pd.read_csv("ventsCleaned")
print(vents.head())
```

```python
#make created at a date-time
vents['created_at'] = pd.to_datetime(vents['created_at'])
```

```python
#sort by user, and then by vent creation time
vents_sorted = vents.sort_values(by=['user_id', 'created_at'])
#add a column for each post's number in the user's post history
vents_sorted['usersPostNum'] = vents_sorted.groupby('user_id').cumcount()
```

```python
#filter by users'second ever posts
second_posts = vents_sorted[vents_sorted['usersPostNum'] == 1].copy()
```

```python
#save csv
second_posts.to_csv("ventsCleaned")
```
