```python
import pandas as pd

vents = pd.read_csv("ventsCleaned")
```

```python
print(vents.head())
```

To calculate the 'virality' a vent had, we will look at reactions and followers. A significantly high number of reactions, when divided by follower number, would count as a highly viral vent. 

Formula: virality = reactions / (followers + 1)

```python
vents["virality"] = vents["reactions"] / (vents["follower_count"] + 1)
print(vents.head())
```

Now we have virality as part of the dataframe, we want to take the top 1/4th most viral vents for our analysis

```python
threshold = vents['virality'].quantile(0.75)

highlyViralVents = vents[vents['virality'] >= threshold]

print(highlyViralVents.head())
print(f"Size of subset: {len(highlyViralVents)}")

import numpy as np

#extra stuff to help SPSS import 
highlyViralVents = highlyViralVents.drop(columns=['Unnamed: 0'])
highlyViralVents.to_csv('highlyviral.csv', index=False, encoding='utf-8-sig')
```
